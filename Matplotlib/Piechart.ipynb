```python
import matplotlib.pyplot as plt

brands=["Oneplus","Apple","Samsung","Nokia","Redimi"]
x=[22,35,30,3,10]
c=["red","silver","blue","pink","orange"]
ex=[0,0,0,0,0.1]

plt.pie(x,labels=brands,colors=c,explode=ex,startangle=180)#explode is used to highlight the neded pie part
plt.savefig("pie.png",facecolor="green")#facecolor means background
plt.show()
```
