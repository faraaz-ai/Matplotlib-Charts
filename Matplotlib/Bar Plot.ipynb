```python
import matplotlib.pyplot as plt

y=[98,67,88,95,88]
x=["Part1","Part2","Part3","Part4","Part5"]

plt.bar(x,y,color="red")#where it changes the color  
plt.xlabel("Parts of Harry Potter")
plt.ylabel("Popularity",fontsize=17)#fontsize is used increse or dec the size of font
plt.title("Popularity of Different parts of Harry Potter")
plt.show()
```
